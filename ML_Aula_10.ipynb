##União das tabelas

- Inicio do processo de união das 9 tabelas disponibilizadas com a finalidade de produzir uma tabela resultante que possua os elementos mais importantes para a determinação do review_score. No primeiro merge realizado, unimos a tabela de clientes com as respectivas entregas dos pedidos usando o código individual de cada consumidor como parâmetro.

```python
pd.merge(tabela_cliente, tabela_entrega_pedido, on=['customer_id'],how='inner')
```

1- (Cliente, Entregas)

```python
teste = pd.merge(tabela_cliente, tabela_entrega_pedido, on=['customer_id'], how='inner')
teste = pd.merge(teste, tabela_pedido, on=['order_id'], how='inner')
teste = pd.merge(teste, tabela_pagamento, on=['order_id'], how='inner')
teste = pd.merge(teste, tabela_review, on=['order_id'], how='inner')
teste = pd.merge(teste, tabela_vendedor, on=['seller_id'], how='inner')
teste
```

##Limpeza de Colunas

```python
teste = teste.drop(columns=["customer_unique_id"],axis=1)
teste = teste.drop(columns=["customer_city"],axis=1)
teste = teste.drop(columns=["customer_state"],axis=1)
teste = teste.drop(columns=["order_status"],axis=1)
teste = teste.drop(columns=["order_purchase_timestamp"],axis=1)
teste = teste.drop(columns=["order_approved_at"],axis=1)
teste = teste.drop(columns=["order_delivered_carrier_date"],axis=1)
teste = teste.drop(columns=["order_delivered_customer_date"],axis=1)
teste = teste.drop(columns=["order_estimated_delivery_date"],axis=1)
teste = teste.drop(columns=["shipping_limit_date"],axis=1)
teste = teste.drop(columns=["review_creation_date"],axis=1)
teste = teste.drop(columns=["review_answer_timestamp"],axis=1)
teste = teste.drop(columns=["seller_city"],axis=1)
teste = teste.drop(columns=["seller_state"],axis=1)
teste = teste.drop(columns=["review_comment_title"],axis=1)
teste = teste.drop(columns=["review_comment_message"],axis=1)
```

##Import metodos de árvore de decisão

```python
from sklearn.tree import DecisionTreeClassifier, export_graphviz
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

- Para simplificar os dados e evitar criar um dummy com esse dataframe, removemos todos os elementos não-numéricos para que o modelo seja capaz de realizar a execução. Solução encotrada para simplificar os atributos do tipo "Objeto" para tipos numéricos.

```python
df['Cliente'] = df['Cliente'].str.replace(r'\D','')
df['Pedido'] = df['Pedido'].str.replace(r'\D','')
df['Produto'] = df['Produto'].str.replace(r'\D','')
df['Vendedor'] = df['Vendedor'].str.replace(r'\D','')
df['ID_Review'] = df['ID_Review'].str.replace(r'\D','')
df
```

- Realizamos o procedimento de remoção dos elementos não-numéricos para todos as colunas do tipo objeto com exceção do tipo de pagamento pois o tipo de pagamento se resume a poucas opções. Dessa forma, usamos a função get_dummies apenas para o tipo de pagamento.
Portanto, a coluna Tipo_pagamento se divide em quatro colunas com lógica booleana. As novas colunas são: Tipo_pagamento_boleto, Tipo_pagamento_credit_card, Tipo_pagamento_debit_card, Tipo_pagamento_voucher.

```python
result_df = pd.get_dummies(df, columns=['Tipo_pagamento'])
result_df
```

##Criação de reserva para possíveis Conclusões

```python
reserva = result_df.copy()
reserva
```

- Eliminando todas as linhas com nível 4 ou nível 5 de satisfação. Dessa forma, temos um dataframe com todos os dados, e um apenas com
dados classificados com nível 3,2 ou 1 ou seja, mediano, ruim ou péssimo. (elementos que apresentam nível de insatisfação interessante
para análise).

```python
reserva = reserva.drop(reserva[reserva.Score > 3].index)
reserva
```

##Processo de separação de treino/teste

###Proporções estabelecidas
**70% Treino**

**30% Teste**

```python
x_train, x_teste, y_train, y_teste = train_test_split(result_df.drop('Score', axis=1),result_df['Score'], test_size=0.3)
```

##Treinamento

```python
cls = cls.fit(x_train, y_train)
```

Portanto, com base nos resultados obtidos, o dono do marketplace deve se atentar aos seguintes parâmetros de sua logística:



1- Relação entre CEP_Cliente, Frete e CEP_Vendedor

Esses atributos apresentaram importancia direta na nota registrada pelo cliente, dessa forma, algumas razões devem ser consideradas:

- Problemas com a entrega
- Qualidade da entrega (tempo de entrega, comprometimento do produto durante o processo de transporte)
- Alto preço do frete para determinada regiões

2- Produto

O atributo "produto" apresentou importancia direta na insatisfação dos clientes. Portanto, deve-se considerar:

- Qualidade ruim de determinados produtos.
- O produto entrege não apresenta as características do produto anunciado.
-Entrega de produtos errados, problema de logística.

3- Vendedor
O atributo "vendedor" apresentou importancioa direta na insatisfação dos clientes. Portante, deve-se considerar:

- Quanlidade duvidosa do atendimento por parte do vendedor.
-Erro do vendedor em algumas etapas do processo em específico.
